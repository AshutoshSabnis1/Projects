```python
import pandas as pd
import numpy as np
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
```

```python
df=pd.read_csv("car.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
df=df.replace("?",np.nan)
```

```python
df.isnull().sum()[df.isnull().sum()*100/df.shape[0]>40]
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.normalized_losses=df.normalized_losses.astype("float")
df.bore =df.bore.astype("float")
df.stroke =df.stroke.astype("float")
df.horsepower=df.horsepower.astype("float")
df.peak_rpm =df.peak_rpm.astype("float")
df.price=df.price.astype("float")
```

```python
df.normalized_losses=df.normalized_losses.fillna(df.normalized_losses.mean())
```

```python
df.bore=df.bore.fillna(df.bore.mean())
```

```python
df.stroke=df.stroke.fillna(df.stroke.mean())
```

```python
df.horsepower=df.horsepower.fillna(df.horsepower.mean())
```

```python
df.peak_rpm=df.peak_rpm.fillna(df.peak_rpm.mean())
```

```python
df.num_of_doors=df.num_of_doors.fillna("four")
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df=df.dropna() # Drop null values from Target variable
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df[df.select_dtypes(include=["object"]).columns]=df[df.select_dtypes(include=["object"]).columns].apply(le.fit_transform)
```

```python
#df=df1
```

```python
from sklearn.model_selection import train_test_split
a_train,a_test=train_test_split(df,test_size=0.2)
```

```python
a_train_x=a_train.iloc[:,0:-1]
a_train_y=a_train.iloc[:,-1]
```

```python
a_test_x=a_test.iloc[:,0:-1]
a_test_y=a_test.iloc[:,-1]
```

```python
from sklearn.linear_model import LinearRegression
le=LinearRegression()
le.fit(a_train_x,a_train_y)

```

```python
Rsquare=le.score(a_test_x,a_test_y)
Rsquare
```

```python
N=a_test_x.shape[0]
K=a_test_x.shape[1]
adj=1-(1-Rsquare)*(N-1)/(N-K-1)
adj
```

```python
pred_train=le.predict(a_train_x)
pred_test=le.predict(a_test_x)
```

```python
err_train=a_train_y-pred_train
err_test=a_test_y-pred_test
```

```python
mse=np.mean(np.square(err_test))
```

```python
mse
```

```python
rmse=np.sqrt(mse)
rmse
```

```python
mape=np.mean(np.abs(err_test*100/a_test_y))
```

```python
mape
```
